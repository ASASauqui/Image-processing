## Libraries

```python
# Determinadas
import numpy as np
import pandas
import matplotlib.pyplot as plt

# Procesamiento de im√°genes
import numpy as np
import cv2
```

## Load Data

```python
# Cargar imagen.
img = cv2.imread('Datasets/Img02.jpg')

# Imagen a escala de grises.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mostrar imagen.
plt.figure()
plt.imshow(img)
plt.show()
```

## Histogram function

```python
def createHistogramArray(img):
    # Sacar la forma de la imagen.
    nrows, ncols, nch = img.shape

    # Crear arreglo vac√≠o de ceros.
    histValues = np.zeros((256,3))

    # Contar pixeles.
    for i in range(nrows):
        for j in range(ncols):
            histValues[img[i,j,0],0] = histValues[img[i,j,0],0] + 1
            histValues[img[i,j,1],1] = histValues[img[i,j,1],1] + 1
            histValues[img[i,j,2],2] = histValues[img[i,j,2],2] + 1
            
    return histValues
```

## Get values

```python
histValues = createHistogramArray(img)
```

## Show histograms

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(range(0,len(histValues)), histValues[:,0], color="red")
plt.subplot(1,3,2)
plt.bar(range(0,len(histValues)), histValues[:,1], color="green")
plt.subplot(1,3,3)
plt.bar(range(0,len(histValues)), histValues[:,2], color="blue")
plt.show()
```
