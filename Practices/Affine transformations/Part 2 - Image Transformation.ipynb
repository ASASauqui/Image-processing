### Libraries

```python
# Determinadas
import numpy as np
import pandas
import matplotlib.pyplot as plt
import copy
import math

# Procesamiento de imágenes
import numpy as np
import cv2
from scipy import ndimage as ndi
```

### Transformation function

```python
def transformation(img, s=[1,1], t=[0,0], r=[0,0], theta=''):
    # Crear matriz de transformación afín.
    mat = np.array([[s[0],r[1],t[1]],[r[0],s[1],t[0]],[0,0,1]])
    
    # Si sí se puso una theta.
    if theta != '':
        # Se aplica la correspondiente multiplicación de matrices para hacer la rotación.
        mat = mat @ np.array([[np.cos(theta),np.sin(theta),0],[np.sin(theta),-np.cos(theta),0],[0,0,1]]) @ np.array([[s[0],r[1],-t[1]],[r[0],s[1],-t[0]],[0,0,1]])
   
    # Aplicar transformación.
    img_T = ndi.affine_transform(img, mat)
    
    return img_T
```

### Load Data

```python
# Cargar imagen.
img = cv2.imread('Datasets/Face.jpg')

# Imagen a escala de grises.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Obtener forma de la imagen.
h, w = img.shape

# Mostrar imagen.
plt.figure()
plt.imshow(img, cmap="gray")
plt.show()
```

### Scale

```python
# Variables.
s= [0.5, 0.5]

# Obtener transformación.
img_T = transformation(img.copy(), s=s)

# Mostrar imagen.
plt.figure()
plt.imshow(img_T, cmap="gray")
plt.show()
```

### Traslation

```python
# Variables.
t = [500, 0]

# Obtener transformación.
img_T = transformation(img.copy(), t=t)

# Mostrar imagen.
plt.figure()
plt.imshow(img_T, cmap="gray")
plt.show()
```

### Rotation

```python
# Variables.
theta = np.pi/3

# Obtener transformación.
img_T = transformation(img.copy(), theta=theta)

# Mostrar imagen.
plt.figure()
plt.imshow(img_T, cmap="gray")
plt.show()
```

### Combination

```python
# Variables.
theta = np.pi/4
s = [0.7,0.7]
t = [w/2,h/2]

# Obtener transformación.
img_T = transformation(img.copy(), s=s, t=t, theta=theta)

# Mostrar imagen.
plt.figure()
plt.imshow(img_T, cmap="gray")
plt.show()
```
