### Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Plot function

```python
def plot_v_dif(v_vector, w_vector):
    v = [np.float(v_vector[0]), np.float(v_vector[1])]
    w = [np.float(w_vector[0]), np.float(w_vector[1])]

    array = np.array([[0, 0, v[0], v[1]], 
                     [0, 0, w[0], w[1]]])

    X, Y, V, W = zip(*array)
    plt.figure()
    plt.ylabel('Y-axis')
    plt.xlabel('X-axis')
    ax = plt.gca()
    ax.annotate(f'v({v[0]},{v[1]})', (v[0],v[1]),fontsize=14)
    plt.scatter(v[0],v[1], s=10,c='red')
    ax.annotate(f'w({w[0]},{w[1]})', (w[0],w[1]),fontsize=14)
    plt.scatter(w[0], w[1], s=10,c='blue')
    ax.quiver(X, Y, V, W, angles='xy', scale_units='xy',color=['r','b'],scale=1)
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])

    plt.grid()
    plt.draw()
    plt.show()
```

### Transform function

```python
def transformVector(vector, a, b, c, d, tx, ty):
    # Crear transformación afín.
    A =np.matrix([[a, b], [c, d]])
    t_vector =np.matrix([[tx],[ty]])

    # Vector de transformación afín.
    w_vector = np.matmul(A,v_vector) + t_vector
    
    # Plotear vector creado con el vector de muestra.
    plot_v_dif(vector, w_vector)
```

### Scaling

```python
# Variables.
vx = 2
vy = 4

# Vector a utilizar.
v_vector = np.matrix([[vx],[vy]])

# Aplicar transformación.
transformVector(v_vector, 2, 0, 0, 2, 0, 0)
```

### Traslation

```python
# Variables.
vx = 2
vy = 4

# Vector a utilizar.
v_vector = np.matrix([[vx],[vy]])

# Aplicar transformación.
transformVector(v_vector, 1, 0, 0, 1, -2, 5)
```

### Rotation

```python
# Variables.
vx = 8
vy = 0
theta = 90

# Funciones trigonométricas.
s = np.sin(np.radians(theta))
c = np.cos(np.radians(theta))

# Vector a utilizar.
v_vector = np.matrix([[vx],[vy]])

# Aplicar transformación.
transformVector(v_vector, c, s, s, -c, 0, 0)
```

### Combination

```python
# Variables.
vx = 8
vy = 0
theta = 90

# Funciones trigonométricas.
s = 0.5*np.sin(np.radians(theta))
c = 0.5*np.cos(np.radians(theta))

# Vector a utilizar.
v_vector = np.matrix([[vx],[vy]])

# Aplicar transformación.
transformVector(v_vector, c, s, s, -c, -1, 0)
```
